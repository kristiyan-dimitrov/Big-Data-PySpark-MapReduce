```python
# Testing EWMA
def ewma(alpha, n):
    
    res = 0
    
    for ii in range(n):
        res += alpha * ( (1-alpha)**ii )
    
    return res
    
```

```python
# Correct!
ewma(.2, 3)
```

```python
## Testing EWMA with a range of values
def ewma_2(alpha, lst):
    
    res = 0
    
    for ii in range(len(lst)):
        res += alpha * ( (1-alpha)**ii ) * lst[ii]
    
    return res
```

```python
# Correct!
ewma_2(.2, [1,1,1])
```
